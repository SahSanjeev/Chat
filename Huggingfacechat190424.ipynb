```python
import os
import chainlit as cl
from langchain import HuggingFaceHub, PromptTemplate ,LLMChain
```

```python
from getpass import getpass
HUGGINGFACEHUB_API_TOKEN = getpass()
os.environ['HUGGINGFACEHUB_API_TOKEN'] = HUGGINGFACEHUB_API_TOKEN
```

```python
model_id = 'gpt2-medium'

conv_model = HuggingFaceHub(huggingfacehub_api_token= os.environ['HUGGINGFACEHUB_API_TOKEN'],
                            repo_id = model_id,
                            model_kwargs ={"temperature":0.8, "max_new_tokens":200}

                            )
```

```python
template = """ You are a helpful AI assistant that makes stories by completing the query provided by the user
{query}"""

prompt =PromptTemplate(template = template , input_variables= ['query'])
```

```python
conv_chain =LLMChain(llm =conv_model,
                     prompt = prompt,
                     verbose =True)
```

```python
print(conv_chain.run("Once upon a time in India"))
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
```

```python
model_name = "facebook/blenderbot-400M-distill"
```

```python
model =AutoModelForSeq2SeqLM.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
! pip install flask-ngrok
```

```python
conversation_history = []
```

```python
history_string = "\n".join(conversation_history)
history_string
```

```python
#input_text = "hello, how are you doing ?"
input_text = "i'm doing very well too, lol. Yes, I love tennis, what about you ?"
```

```python
inputs = tokenizer.encode_plus(history_string, input_text, return_tensors="pt")
inputs
```

```python
#tokenizer.pretrained_vocab_files_map
```

```python
outputs = model.generate(**inputs)
outputs
```

```python
response = tokenizer.decode(outputs[0], skip_special_tokens=True).strip()
response
```

```python
conversation_history.append(input_text)
conversation_history.append(response)
conversation_history
```

```python
while True:
    # Create conversation history string
    history_string = "\n".join(conversation_history)

    # Get the input data from the user
    input_text = input("> ")

    # Tokenize the input text and history
    inputs = tokenizer.encode_plus(history_string, input_text, return_tensors="pt")

    # Generate the response from the model
    outputs = model.generate(**inputs)

    # Decode the response
    response = tokenizer.decode(outputs[0], skip_special_tokens=True).strip()

    # Add interaction to conversation history
    conversation_history.append(input_text)
    conversation_history.append(response)
```
